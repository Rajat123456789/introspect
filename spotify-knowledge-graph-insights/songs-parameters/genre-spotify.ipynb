```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('spotify-history-processed.csv')
df.head()

```

```python
#remove spotify:track: from url column and rename to track_id
df['track_id'] = df['url'].str.replace('spotify:track:', '')
df.head()

#remove url column
df = df.drop('url', axis=1)
```

```python
df.head()
```

```python
ds = pd.read_csv('spotify.csv')
```

```python
ds.head()
```

```python
join_df = pd.merge(df, ds, on='track_id', how='left')
join_df.head()
join_df.to_csv('spotify-history-processed.csv', index=False)
```
