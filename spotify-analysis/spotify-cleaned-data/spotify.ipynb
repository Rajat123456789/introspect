```python


# Define the file path
file_path = 'spotify-history-cleaned.csv'

def process_csv(file_path):
    df = pd.read_csv(file_path)
    
    # Ensure 'startDate' and 'end_time' are in datetime format
    df['startDate'] = pd.to_datetime(df['startDate'])
    df['end_time'] = pd.to_datetime(df['end_time'])
    
    # Round down timestamps to nearest 10-minute interval
    df['interval_start'] = df['startDate'].dt.floor('10T')
    df['interval_end'] = df['end_time'].dt.floor('10T')
    
    # Keep the latest row for each interval
    df = df.sort_values(by=['interval_start', 'startDate'], ascending=[True, False])
    df = df.drop_duplicates(subset=['interval_start'], keep='first')
    
    df = df.sort_values(by=['interval_end', 'end_time'], ascending=[True, False])
    df = df.drop_duplicates(subset=['interval_end'], keep='first')
    
    # Replace startDate and end_time with interval_start and interval_end
    df['startDate'] = df['interval_start']
    df['end_time'] = df['interval_end']
    df = df.drop(columns=['interval_start', 'interval_end'])
    
    return df

# Process the single CSV file
processed_df = process_csv(file_path)

# Save processed file
processed_df.to_csv(file_path, index=False)
print(f'Processed: {file_path}')
```
