```python
import pandas as pd
import numpy as np

yt = pd.read_csv('/Users/rajatsharma/10-feb-introspect-proj/combining/combined-youtube-data.csv', encoding='ISO-8859-1')
hth = pd.read_csv('/Users/rajatsharma/10-feb-introspect-proj/combining/combined-health-data.csv', encoding='ISO-8859-1')
```

```python
#drop all columns except for the timestamp and steps_10min
hth = hth[['timestamp', 'Steps', 'Heart Rate']]
```

```python
hth.head()
```

```python
#rename Watched At column to timestamp in yt dataframe
yt = yt.rename(columns={'Watched At': 'timestamp'})
```

```python
#combine the two dataframes with the timestamp as the key

df = pd.merge(yt, hth, on='timestamp', how='outer')

df.head()
```

```python
#save the combined dataframe to a csv file
df.to_csv('combined_data.csv', index=False)
```
