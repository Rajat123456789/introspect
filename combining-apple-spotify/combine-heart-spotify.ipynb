```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
spotify = pd.read_csv('spotify-history-processed.csv')
audioExposure = pd.read_csv('AudioExposureEvent.csv')
heartRate = pd.read_csv('HeartRate.csv')
steps = pd.read_csv('StepCount.csv')
```

```python
heartRate.head()
```

```python
audioExposure.head()
```

```python
steps.head()
```

```python
#drop column name after 6th column
heartRate = heartRate.drop(heartRate.columns[6:], axis=1)
```

```python
audioExposure = audioExposure.drop(audioExposure.columns[6:], axis=1)
steps = steps.drop(steps.columns[6:], axis=1)
```

```python
heartRate.head()
```

```python
steps.head()
```

```python
#rename column end_time in spotify to startDate
spotify = spotify.rename(columns={'end_time': 'startDate'})
```

```python
spotify.head()
```

```python
#join spotify and heartRate on startDate
spotifyHeartRate = spotify.merge(heartRate, on='startDate', how='inner')

spotifyHeartRate.head()
```

```python
spotifyHeartRate.shape
```

```python
#join spotify and stepCount on startDate
spotifyStepCount = spotify.merge(steps, on='startDate', how='inner')

spotifyStepCount.head()
```

```python
spotifyStepCount.shape
```

```python
#save spotifyHeartRate and spotifyStepCount to csv in a different folder

spotifyHeartRate.to_csv('spotifyHeartRate.csv', index=False)
spotifyStepCount.to_csv('spotifyStepCount.csv', index=False)
```
